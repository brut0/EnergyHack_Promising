```python
import time as tm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 50)
```

```python
PATH = './data/'

# Load data
df_fiders = pd.read_excel(PATH+'Графики.xlsx', index_col=None, usecols=['Час', 'График 1, %', 'График 2, %'])
```

```python
df_fiders
```

```python
fig = plt.figure(figsize=(18,6))
plt.plot(df_fiders['Час'], df_fiders['График 1, %'], label='Деловой центр')
plt.plot(df_fiders['Час'], df_fiders['График 2, %'], label='Спальный район')
plt.plot(df_fiders['Час'], (df_fiders['График 1, %']+df_fiders['График 2, %'])/2, label='Сумма')
step = 8
plt.xticks(range(min(df_fiders['Час'])+1, max(df_fiders['Час'])+step+1, step))
plt.ylabel('%')
plt.legend(loc="lower right")
plt.show()
```

```python
df_hours = pd.read_excel(PATH+'Графики.xlsx', sheet_name='суточные графики', index_col=None)
```

```python
df_hours['Время'] = pd.date_range("00:00", "23:30", freq="30min").time
```

```python
df_hours
```

```python
pd.plotting.register_matplotlib_converters()

fig = plt.figure(figsize=(18,6))
for c in df_hours.columns[1:]:
    plt.plot(df_hours['Время'], df_hours[c], label=f'{c}')
plt.ylabel('%')
plt.legend(loc="lower right")
plt.show()
```
